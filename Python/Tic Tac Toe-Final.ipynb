# Play Tic-Tac-Toe

```python
import pandas as pd
import numpy as np
```

```python
#Define valid inputs for selecting board spaces
available = ["A1", "B1", "C1",
            "A2", "B2", "C2",
            "A3", "B3", "C3",]

#Initialize the list of chosen spots for both players
xspots = []
ospots = []

#Create list of possible winning sets
winner = [["A1", "B1", "C1"],
          ["A2", "B2", "C2"],
         ["A3", "B3", "C3"],
         ["A1", "A2", "A3"],
         ["B1", "B2", "B3"],
         ["C1", "C2", "C3"],
         ["A1", "B2", "C3"],
         ["A3", "B2", "C1"]]

#Create board for tic-tac-toe and print blank board to start
board = pd.DataFrame(index = [1, 2, 3], columns = ["A", "B", "C"])
print(board.replace(to_replace=np.nan, value=" "))

#Playing game loop:
playing = True
while (playing == True):
    
    #Player X goes
    position = input("Choose a position, Player X: ")
    if position in available:
        available.remove(position)
        xspots.append(position)
        
    #If position invalid, prompt player to choose valid position:
    else:
        invalid = True
        
        #Check both types of invalid position
        while(invalid):
            if position in xspots or position in ospots:
                position = input("Previous position already chosen. Choose a valid position, Player X: ")
            else:   
                position = input("Previous position invalid. Choose a valid position, Player X: ")
            if position in available:
                available.remove(position)
                xspots.append(position)
                invalid = False
            else:
                invalid = True
    
    #Show current spaces marked and current spaces available
    datloc = list(position)
    board.loc[int(datloc[1]),datloc[0]] = "X"
    print(board.replace(to_replace=np.nan, value=" "))
    
    #Check if Player X won or tie:
    for i in range(0, 8):
        if set(winner[i]).issubset(set(xspots)):
            print("Player X Wins the Game!")
            playing = False
            break
        else:
            playing = True
    if playing == True and len(available) == 0:
            print("Tie Game!")
            playing = False
            break
    if playing == False:
        break
    
    
    #Player O goes
    position = input("Choose a position, Player O: ")
    if position in available:
        available.remove(position)
        ospots.append(position)
    #If position invalid, prompt player to choose valid position:
    else:
        invalid = True
        
        #Check both types of invalid position
        while(invalid):
            if position in xspots or position in ospots:
                position = input("Previous position already chosen. Choose a valid position, Player O: ")
            else:   
                position = input("Previous position invalid. Choose a valid position, Player O: ")
            if position in available:
                available.remove(position)
                ospots.append(position)
                invalid = False
            else:
                invalid = True
    
    #Show current spaces marked and current spaces available
    datloc = list(position)
    board.loc[int(datloc[1]),datloc[0]] = "O"
    print(board.replace(to_replace=np.nan, value=" "))
    
    #Check if Player O won or tie:
    for i in range(0, 8):
        if set(winner[i]).issubset(set(ospots)):
            print("Player O Wins the Game!")
            playing = False
            break
        else:
            playing = True
    if playing == True and len(available) == 0:
            print("Tie Game!")
            playing = False
            break
    if playing == False:
        break
    
    
        
print("Game Over.")
    
```
