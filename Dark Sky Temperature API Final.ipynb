```python
import pandas as pd
import csv
from forecastiopy import *
```

```python
api_key = 'f317a600493f0218b6a2dc86e29e48e3'

#Cities to Query [City, Latitude, Longitude]
loc = [
    ["Anchorage, Alaska", 61.2181, -149.9003],
    ["Buenos Aires, Argentia", -34.6037, -58.3816],
    ["São José dos Campos, Brazil", -23.2237, -45.9009],
    ["San José, Costa Rica", 9.9281, -84.0907],
    ["Nanaimo, Canada", 49.1659, -123.9401],
    ["Ningbo, China", 29.8683, 121.5440],
    ["Giza, Egypt", 30.0131, 31.2089],
    ["Mannheim, Germany", 49.4875, 8.4660],
    ["Hyderabad, India", 17.3850, 78.4867],
    ["Tehran, Iran", 35.6892, 51.3890],
    ["Bishkek, Kyrgyzstan", 42.8746, 74.5698],
    ["Riga, Latvia", 56.9496, 24.1052],
    ["Quetta, Pakistan",30.1798, 66.9750],
    ["Warsaw, Poland", 52.2297, 21.0122],
    ["Dhahran, Saudia Arabia", 26.2361, 50.0393],
    ["Madrid, Spain", 40.4168, -3.7038],
    ["Oldham, United Kingdom", 53.5409, -2.1114]
]
```

```python
data = []

#Add headers
data.append(["City", "Min 1", "Max 1", 
                            "Min 2", "Max 2",
                            "Min 3", "Max 3",
                            "Min 4", "Max 4",
                            "Min 5", "Max 5",
                            "Min Avg", "Max Avg"])

#Fill in the Data
for city in loc:
    #Initialize min and max temp list
    mins = []
    maxs = []
    
    #Load the forecast object
    weather = ForecastIO.ForecastIO( api_key, units='si', latitude=city[1], longitude=city[2] )
    daily = FIODaily.FIODaily( weather )
        
    #Forecast temps
    for day in range( 2, 7 ):
        temps = daily.get(day)
        mintemp = temps[ 'temperatureMin' ]
        maxtemp = temps[ 'temperatureMax' ]
        mins.append(mintemp)
        maxs.append(maxtemp)
   
    #Add values to the dataset
    data.append([city[0],
    "{0:.2f}".format(mins[0]), #min1
    "{0:.2f}".format(maxs[0]), #max1
    "{0:.2f}".format(mins[1]), #min2
    "{0:.2f}".format(maxs[1]), #max2
    "{0:.2f}".format(mins[2]), #min3
    "{0:.2f}".format(maxs[2]), #max3
    "{0:.2f}".format(mins[3]), #min4
    "{0:.2f}".format(maxs[3]), #max4
    "{0:.2f}".format(mins[4]), #min5
    "{0:.2f}".format(maxs[4]), #max5
    "{0:.2f}".format(sum(mins)/len(mins)), #avgmin
    "{0:.2f}".format(sum(maxs)/len(maxs)) #avgmax
                ])
```

```python
with open('temp.csv', 'w', newline='') as csvtemp:
    writer = csv.writer(csvtemp)
    writer.writerows(data)
csvtemp.close()
```
